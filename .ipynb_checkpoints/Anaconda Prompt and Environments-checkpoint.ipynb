# Anaconda Prompt and Environments

Open Anaconda prompt from the start menu. It's a command line interface.

## Opening Jupyter Notebook

A Jupyter notebook can then be opened by typing `jupyter notebook` in the command line.

## Environments
Using environments when working in python allows for efficient and consistent use of packages, and versions of packages, for various projects. It also make it easy to change to different versions of python depending on the project. 

### Create environment

To create an enironment, type the following in the command line (where `myenv` is the name of the environment)
```
conda create --name myenv
```

To create an environment with a specific version of python -
```
conda create -n myenv python=3.6
```

To activate an environment, type
```
conda activate myenv
```

The current environment is shown in `()` brackets at the beginning of the command line - 

![current environment](pictures-for-notebooks/current-environment.png)

To deactivate an environment, type
```
conda deactivate
```


### Install packages in environment

To view a list of packages installed in an environment, type the following if the environment is not activated - 
```
conda list -n myenv
```
or the following if the environment is activated -
```
conda list
```

To install new packages if the environment is not activated (where mypackage is the package to be installed) -
```
conda install -n myenv mypackage
```
or with environment activated - 
```
conda install mypackage
```

Alternatively, `pip install` can be used. Pip is a package manager in Python to help with managing libraries and dependancies that are not part of the standard python library. Pip should be included as a standard package in python. To check the version, run
```
pip --version
```
Pip has a number of functions, which can be viewed by typing
```
pip help
```
One of them is of course `install` which can be used like so -

```
pip install mypackage
```

### Deleting an environment

An environment can be removed by typing 
```
conda remove --name myenv --all
```
